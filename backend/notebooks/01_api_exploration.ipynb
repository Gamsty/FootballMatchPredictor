```python
import requests
import json
from dotenv import load_dotenv
import os

# Load API key
load_dotenv()
API_KEY = os.getenv('FOOTBALL_API_KEY')

# Set up headers
headers = {'X-Auth-Token': API_KEY}
base_url = 'https://api.football-data.org/v4'

# Get available competitions
competitions_url = f'{base_url}/competitions'
response = requests.get(competitions_url, headers=headers)
competitions = response.json()

print("Available competitions:")
for comp in competitions['competitions'][:10]: # Show first 10
    print(f"ID: {comp['id']}, Name: {comp['name']}, Code: {comp['code']}")

# Test getting matches
# Premier League ID is 2021 and get matches from 2023 season
matches_url = f'{base_url}/competitions/2021/matches'
params = {'season': 2023}

response = requests.get(matches_url, headers=headers, params=params)
matches = response.json()

print(f"\nTotal matches: {matches['resultSet']['count']}")
print("\nFirst match:")
print(json.dumps(matches['matches'][0], indent=2))

# Look at what data is availiable
first_match = matches['matches'][0]

print("Match data fields:")
print(f"- Home Team: {first_match['homeTeam']['name']}")
print(f"- Away Team: {first_match['awayTeam']['name']}")
print(f"- Date: {first_match['utcDate']}")
print(f"- Status: {first_match['status']}")
print(f"- Score: {first_match['score']['fullTime']}")
```
